```python
import pandas as pd

# Create a simple gene expression dataset
data = {
    "Gene": ["TP53", "BRCA1", "EGFR", "MYC", "PTEN"],
    "Cancer_Expression": [8.5, 7.2, 9.1, 8.9, 6.4],
    "Normal_Expression": [3.2, 4.1, 4.8, 5.0, 3.9]
}

df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.info()
```

```python
df["Expression_Difference"] = df["Cancer_Expression"] - df["Normal_Expression"]
```

```python
df.sort_values(by = "Expression_Difference", ascending=False)
```

```python
import numpy as np

np.mean(df["Cancer_Expression"])
```

```python
np.max(df["Expression_Difference"])
```

## Interpretation

In this analysis, gene expression levels were compared between cancer and normal samples. The results indicate that genes such as EGFR and MYC show higher expression in cancer samples, suggesting their possible involvement in cancer-related biological processes. This project demonstrates basic gene expression analysis using Python.
